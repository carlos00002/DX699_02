# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='ME')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='ME')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt

# Months
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

# Approximate values from the figure
received = [165,190,250,160,195,170,140,202,160,139,149,177]
processed = [165,190,250,160,195,170,135,156,126,104,124,140]

plt.figure(figsize=(10,5))

# Plot the lines
plt.plot(months, received, linewidth=3, label="Received")
plt.plot(months, processed, linewidth=3, label="Processed")

# Label only the last months (Aug–Dec) like the book
for i in range(7,12):
    plt.scatter(months[i], received[i])
    plt.text(i, received[i]+5, str(received[i]), ha='center')

    plt.scatter(months[i], processed[i])
    plt.text(i, processed[i]-10, str(processed[i]), ha='center')

# Styling similar to the book
plt.ylim(0,300)
plt.title("Received vs Processed")
plt.xlabel("")
plt.ylabel("")
plt.grid(axis="y", alpha=0.3)

# Put labels near the lines instead of a legend
plt.text(11.1, received[-1], "Received", va="center")
plt.text(11.1, processed[-1], "Processed", va="center")

plt.tight_layout()
plt.show()
```
